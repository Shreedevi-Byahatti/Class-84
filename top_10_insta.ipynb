Top 10 Influencer based on followers

**Question Find the comparison of posts that are posted by top 10 influencers of the world.**

To solve the question we have to make list of the top 10 influencers. Top 10 influencers are those whose followers are more. So we sort top 10 influencers as per their followers. Hereâ€™s code for this:


```python
value = insta_data['Followers'].sort_values(ascending=False)[:10]
value
```

This line of code filters 'insta_data' to select rows where the 'Followers' column is greater than or equal to the 10th highest follower count value, creating a 'top_10' DataFrame of the top Instagram influencers.


```python
top_10 = insta_data[insta_data['Followers']>=value[9]]
top_10
```

This code creates a pie chart displaying the number of posts by the top 10 influencers, with their usernames as labels, and sets a specific starting angle and percentage format for the chart's appearance.

```python
plt.figure(figsize=(20,8))
plt.pie(top_10['Posts'],labels=top_10['Username'],startangle=58,autopct='%1.2f%%')
plt.title('No of Posts Top 10 Influencer Posted',size=20)
```

**Question : Comparison between top 10 global influencers and top 10 Indian influencers.**

The following code organizes Instagram influencer data by follower count in descending order, and then it extracts two sets of influencers: the top 10 overall influencers and the top 10 Indian influencers based on their follower counts. This allows for separate analysis and visualization of these two groups.


```python
# Sort the data by followers in descending order to get top influencers
sorted_data = insta_data.sort_values(by='Followers', ascending=False)

# Extract the top 10 overall influencers
top_10_overall = sorted_data.head(10)

# Extract the top 10 Indian influencers
top_10_indian = sorted_data[sorted_data['Country'] == 'IN'].head(10)

# Create a bar chart to visualize the comparison
plt.figure(figsize=(12, 6))
plt.bar(top_10_overall['Username'], top_10_overall['Followers'], label='Top 10 Overall')
plt.bar(top_10_indian['Username'], top_10_indian['Followers'], label='Top 10 Indian')
plt.xlabel('Influencer')
plt.ylabel('Number of Followers')
plt.title('Top 10 Instagram Influencers Overall vs. Top 10 Indian Influencers')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()

# Display the graph
plt.show()
```

**Question: How does the provided code use data sorting and visualization to compare the average views and likes of the top 10 overall Instagram influencers?**

This code first sorts an Instagram influencer dataset ('insta_data') by follower count in descending order. Then, it extracts the top 10 overall influencers based on their followers and separately selects the top 10 Indian influencers by filtering for the 'India' country tag in the dataset.


```python
sorted_data = insta_data.sort_values(by='Followers', ascending=False)

# Extract the top 10 overall influencers
top_10_overall = sorted_data.head(10)

# Extract the top 10 Indian influencers
top_10_indian = sorted_data[sorted_data['Country'] == 'India'].head(10)

# Create a bar chart to visualize the comparison
plt.figure(figsize=(12, 6))

# Plot Followers on the left side of each bar and Likes on the right side
sns.barplot(x='Views Avg.', y='Username', data=top_10_overall, color='skyblue', label='Views Avg.')
sns.barplot(x='Likes Avg.', y='Username', data=top_10_overall, color='lightcoral', label='Likes Avg.')

plt.xlabel('Count')
plt.title('Top 10 Instagram Influencers Overall - Likes and Views',size=20)
plt.legend(loc='upper right')

# Display the graph
plt.show()
```
