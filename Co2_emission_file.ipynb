To get started with Data Visualisation, we will first import the library pandas, matplotlib.pyplot & seaborn. This sets up all the necessary components for CO2 Data Emissions data analysis for visual representation.


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

Here we are loading the dataset.

```python
url = 'https://drive.google.com/uc?id=1mVBXALH-5M83bMzL6I9YeE7-_fZ_P-OW'
```

```python
co2_data=pd.read_csv(url)
co2_data.head()
```

Here we are checking the information of the data say
*   No of columns in the data
*   No of row in the data
*   No of non null values in each row
*   datatype of each row

```python
co2_data.info()
```

```python
co2_data.isnull().sum().max()
```

Now we will drop the rows which have any null value and then reset the index for the remaining data
We can also fill up the values by mean median or mode but here we are only visualizing the data so we will drop it

```python
co2_data.dropna(inplace=True)
```

```python
co2_data.isnull().sum().max()
```

```python
co2_data.reset_index(drop=True, inplace=True)
```

```python
co2_data.head()
```

```python
co2_data.columns
```

As we can see almost all the values in the Indicator Name are same so we will check for the same and drop the column if all the values are same

```python
co2_data['Indicator Name'].value_counts()
```

```python
co2_data['2019'].equals(co2_data['2019.1'])
```

As all the values turn out to be the same as we are using a part of dataset that shows only CO2 emission we will drop this column
also the last column 2019.1 is same as column 2019 so we will drop it

```python
co2_data.drop(['Indicator Name','country_code','2019.1'],axis=1,inplace=True)
co2_data.head()
```

Now we are using Data describe method from pandas to calculate the mean,count,std,min etc. of the columns

```python
co2_data.describe()
```

Now we will start our Data Visualising Step

Now we will plot a bar graph of Regions against the emission in year 2019 and 1990


```python
co2_data.columns
```

```python
co2_data['Total Emission'] = co2_data[['1990', '1991', '1992', '1993', '1994',
       '1995', '1996', '1997', '1998', '1999', '2000', '2001', '2002', '2003',
       '2004', '2005', '2006', '2007', '2008', '2009', '2010', '2011', '2012',
       '2013', '2014', '2015', '2016', '2017', '2018', '2019']].sum(axis=1)
```

```python
co2_data.columns
```

```python
co2_data.head()
```

```python
region_total = co2_data.groupby('Region')['Total Emission'].sum().reset_index()
region_total
```

```python
# Create the bar plot using Seaborn
plt.figure(figsize=(10, 6))
sns.barplot(data=region_total, x='Region', y='Total Emission')
plt.xlabel('Region')
plt.ylabel('Total Emission')
plt.title('Total Emission by Region')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.show()
```

Now we will make a pie chart of Those 10 Countries that had the maximum emission in year 2019

```python
top10=co2_data.sort_values(by='2019',ascending=False).head(10)
lab=top10['Country Name']
data1=top10['2019']
plt.figure(figsize=(20,8))
plt.pie(data1,labels=lab,startangle=0,autopct='%1.2f%%')
plt.title('Countries with Maximum Emissions',size=20)
```

Now we will plot a graph Showing 10 Countries with Overall Maximum Emission and 10 countries with the maximum emission in the largest region

```python
# Find the largest region
largest_region = co2_data['Region'].value_counts().idxmax()

# Get the top 10 emitters overall for 2019
top_10_overall = co2_data.sort_values(by='2019', ascending=False).head(10)

# Get the top 10 emitters in the largest region for 2019
top_10_region = co2_data[co2_data['Region'] == largest_region].sort_values(by='2019', ascending=False).head(10)

# Create a bar chart to compare emissions
plt.figure(figsize=(12, 6))
plt.bar(top_10_overall['Country Name'], top_10_overall['2019'], label='Max Overall Emission')
plt.bar(top_10_region['Country Name'], top_10_region['2019'], label=f'Max Emission in {largest_region}')
plt.xlabel('Country')
plt.title('Max Emission Overall vs. Max Emission in a Region')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.show()
```
