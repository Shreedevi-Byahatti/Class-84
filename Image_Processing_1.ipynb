```python
#Activity 1 (Explaining about the uses of Matplotlib )
import matplotlib.pyplot as plt

# Some data points (x and y coordinates)
x = [1, 2, 3, 4, 5]
y = [10, 12, 5, 8, 7]

# Create a plot
plt.plot(x, y)

# Add labels to the axes
plt.xlabel('X-axis')
plt.ylabel('Y-axis')

# Add a title
plt.title('My First Matplotlib Plot')

# Show the plot
plt.show()
```

```python
#Third Activity - Learning about the Numpy
import numpy as np

# Create a 1D NumPy array of numbers from 0 to 4
my_array = np.arange(5)

# Square each element in the array
squared_array = my_array ** 2

# Calculate the sum of the squared values
sum_of_squares = squared_array.sum()

print("Original Array:", my_array)
print("Squared Array:", squared_array)
print("Sum of Squares:", sum_of_squares)
```
