```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
url = 'https://drive.google.com/uc?id=1id___NYaSi1FYR8N_12N-hCFtB2pJJs7'
```

```python
population = pd.read_csv(url)
population
```

```python
population.info()
```

For the simplest graph plotting the population of top 10 most populated countries in 2022



Locating the top 10 countries by sort value method,using ascending=false so that to get highest values on top

```python
top_10 = population.sort_values(by=['2022 Population'],ascending=False).head(10)
top_10
```

code for the  bar graph


```python
plt.figure(figsize=(12,6))
plt.plot(top_10['Country/Territory'],top_10['2022 Population'],color='#618264',label='Population'
        ,marker='d',markersize=8, markerfacecolor='red',markeredgecolor='black')
plt.title('Population of the top most populated countries in 2022',size=30,color='#A2C579')
plt.xlabel('Country name',size=18)
plt.ylabel('population in 2022 in cr',size=18)
plt.grid(alpha=0.5)
plt.legend()
plt.show()
```

for the second graph comparing the population of india in the given dataset


first of all storing the data as an array in a variable from the previously created top_10 variable using iloc method

```python
population_series = top_10.iloc[1,5:12]
```

```python
population_series
```

code for the  chart

```python
plt.figure(figsize=(15,6))
plt.bar(population_series.index,population_series,width=0.5,color='#c9e52f' )
plt.title('Change in the population of India in multiple of cr',size=30,color='#7C9D96')
plt.xlabel('Year',size=24)
plt.ylabel('population',size=24)
plt.show()
```

for the third chart comparing the difference of population of china and india during the given years and changing the background color of the graph

```python
population_series = top_10.iloc[:2,5:12]
```

```python
population_series
```

claculating the difference

```python
diff = population_series.iloc[0,::]-population_series.iloc[1,::]
```

```python
diff
```

```python
plt.figure(figsize=(12,6))
plt.plot(diff.index,diff,color='blue',marker='d',markersize=8,
         markerfacecolor='lavender',markeredgecolor='black')
plt.bar(diff.index,diff,width=0.4,color='red',label='difference')
plt.title('Difference between the population of China and India ',size=24,color='lime')
plt.xlabel('years',size=24,color='#3A614B')
plt.ylabel('difference b/w population',size=24,color='#3A614B')
plt.show()
```
