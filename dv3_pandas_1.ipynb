Pandas Series from a list

```python

student= ['Rick','Bob','Micky','John']
print(student)
print(type(student))
```

```python
# Importing the pandas library (usually it is imported as "pd" alias)
import pandas as pd
student= ['Rick','Bob','Micky','John']
s1 = pd.Series(student)

print(s1)
print(type(s1))
```

Pandas Series from a nested list

```python
import pandas as pd
student= [['Rick','Bob'],['Micky','John']]
s1 = pd.Series(student)

print(s1)
print(type(s1))
```

Pandas DataFrame from a nested list

```python
student = [['Rick', 5, "Male"]]

# Creating a Pandas DataFrame
df = pd.DataFrame(student)

# Printing the DataFrame
df
```

```python
# Importing the pandas library (usually it is imported as "pd")
import pandas as pd

# Creating a nested list and initialising it
student = [['Rick', 5, "Male"],
           ['Bob', 7, "Male"],
           ['Micky', 8, "Female"],
           ['John', 3, "Male"]]

# Creating a Pandas DataFrame
df = pd.DataFrame(student)

# Printing thetype of DataFrame
print(type(df))

# Printing the DataFrame
df
```

**setting column names** columns=['Name', 'Grade', 'Sex']

```python
# Importing the pandas library (usually it is imported as "pd")
import pandas as pd

# Creating a nested list and initialising it
student = [['Rick', 5, "Male"],
           ['Bob', 7, "Male"],
           ['Micky', 8, "Female"],
           ['John', 3, "Male"]]

# Creating a Pandas DataFrame
df = pd.DataFrame(student, columns=['Name', 'Grade', 'Sex'])

# Printing the DataFrame
df
```

 **loc : location**



```python
#refer to the row index:
print(df1.loc['c'])
print()
print(type(df1.loc['c']))

```

```python
#refer to the row index:
print(df1.loc[['c','d']])
print()
print(type(df1.loc[['c','d']]))
```

```python

print(df1.iloc[0])
print()
print(type(df1.iloc[0]))
```
