```python
# importing the module
import pandas as pd

# creating the DataFrame
df1 = pd.DataFrame({'ID': {0: 501, 1: 502, 2: 503,
                                 3: 504, 4: 505, 5: 506},

                          'Name': {0: 'Alex', 1: 'Chalsea',2: 'Chris',
                                   3: 'Eli',4: 'Mousami', 5: 'Radha'},

                          'Marks': {0: 89, 1: 97, 2: 45,
                                    3: 78, 4: 56, 5: 76},

                          'Grade': {0: 'B', 1: 'A', 2: 'F',
                                    3: 'C', 4: 'E', 5: 'C'}})

df1
```

```python
D2={"categories" : ['Housing',  'Utilities',  'Transportation',  'Groceries',
                    'Entertainment',  'Healthcare',  'Education',  'Miscellaneous'],
    "expenses" : [1200, 300, 400, 500, 200, 150, 250, 100]}
df2 = pd.DataFrame(D2)
df2
```

**creating the excel & csv file from Data Frame**

```python
# saving df in the excel file
df1.to_excel('df1.xlsx')
print('DataFrame is written to Excel File successfully.')
```

Writing **Multiple DataFrames to Different Sheets:**
You can write multiple DataFrames to different sheets within the same Excel file using the ExcelWriter class.

```python
# Create an ExcelWriter object
writer = pd.ExcelWriter('multiple_sheets.xlsx')

# Write DataFrames to different sheets
df1.to_excel(writer, sheet_name='Sheet1', index=False)
df2.to_excel(writer, sheet_name='Sheet2', index=False)

# Save and close the ExcelWriter
writer.save()
```

```python
# saving df in the csv file
df1.to_csv("df1.csv",index=False)
print('DataFrame is written to csv File successfully.')
```
