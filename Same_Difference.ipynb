```python
def compute():
	LIMIT = 10**6
	solutions = [0] * LIMIT
	for m in range(1, LIMIT * 2):
		for k in range(m // 5 + 1, (m + 1) // 2):
			temp = (m - k) * (k * 5 - m)
			if temp >= LIMIT:
				break
			solutions[temp] += 1

	ans = solutions.count(10)
	return str(ans)

compute()
```

```python
def find_pythagorean_triplet(target_sum):
    for a in range(1, target_sum):
        for b in range(a, target_sum):
            c = target_sum - a - b
            if a**2 + b**2 == c**2:
                return a, b, c
    return None

def main():
    target_sum = 1000
    triplet = find_pythagorean_triplet(target_sum)

    if triplet is not None:
        a, b, c = triplet
        product = a * b * c
        print("Pythagorean triplet:", a, b, c)
        print("Product of a, b, and c:", product)
    else:
        print("No Pythagorean triplet found for the given target sum.")

main()
```

```python
def proper_divisors_sum(n):
    divisors = [1]
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            divisors.extend([i, n // i])
    # print(sum(set(divisors))) # This line is not needed here
    return sum(set(divisors)) # Add this line to return the sum

def find_amicable_numbers(limit):
    amicable_numbers = set()

    for a in range(2, limit):
        b = proper_divisors_sum(a)
        if a != b and proper_divisors_sum(b) == a:
            amicable_numbers.add(a)
            amicable_numbers.add(b)

    return amicable_numbers

limit = 10000
amicable_numbers = find_amicable_numbers(limit)
result = sum(amicable_numbers)

print("Amicable numbers under", limit, "are:", amicable_numbers)
print("Sum of amicable numbers:", result)
```
