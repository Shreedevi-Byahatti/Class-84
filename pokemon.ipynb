```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Here we use the pandas read_csv file method to read the pokemon.csv file. The meaning of csv is simply Comma seprated file.

gen9_pokemon_stats.csv
This file contains the Types, Stats and Weaknesses of all 103 Pokémon. Each record is a Pokémon.

- Pokemon - Name of the Pokémon [can be used as a Primary Key for other files]
-Type 1 - First type of the Pokémon
-Type 2 - Second type of the Pokémon, can be empty
- HP - Base HP stat of the Pokémon
- Attack - Base Attack stat of the Pokémon
- Defense - Base Defense stat of the Pokémon
- Special Attack - Base Special Attack stat of the Pokémon
- Special Defense - Base Special Defense stat of the Pokémon
- Speed - Base Speed stat of the Pokémon
- Weakness-Type - The damage multiplier applied to the Pokémon when hit by a move of this type. For example, 0.5 means the Pokémon would take halve damage, and 2 would mean the Pokémon takes double damage.

Getting the file from the google drive

```python
url = 'https://drive.google.com/uc?id=1E-ZWPgtjiAyT_315q21CG4Mvkwcy_d2Z'
```

```python
pokemon_data = pd.read_csv(url)
```

Here we use the pandas head method to read the just first five rows of the DataFrame. By defualt it's shows us the first five rows but we can give our own number of rows that we want as an output.

```python
pokemon_data.head()
```

```python
pokemon_data.columns
```

Here we are Ploting the Number of pokemon of each type with using the matplotlib module.
- In the first line we are setting the figure size with the use of figure method and give the attribute figsize as a tupple
- In the 2nd line we are using the seaborn libary which is the another tool to visulaize the data and giving them the data and the column name and ploting the count plot
- In the 3rd line we giving the title of our graph.
- In the 4th and 5th line we are giving the X and Y labels
- In the last line we are using the show method to display the graph

```python
plt.figure(figsize=(14, 8))
sns.countplot(data=pokemon_data,x='Type 1')
plt.title('Number of pokemons of each type',size=20)
plt.xlabel('Type of pokemon')
plt.ylabel('Count')
plt.show()
```

The code calculates the mean values of attack, defense, HP, special attack, special defense, and speed stats for each Pokémon type (Type 1) by grouping the data using the 'type1' column

```python
df = pokemon_data.groupby(by='Type 1').mean()[['Attack','Defense','HP','Special Attack', 'Special Defense', 'Speed']]
```

Creating a bar plot to show average attributes of each type of Pokémon

- Using the 'plot' function to create a bar plot
- Adjusting the rotation angle of the x-axis labels for better readability
- Setting a descriptive title for the plot with a larger font size
- Adding a label to describe the x-axis (Type of Pokémon)
- Adding a label to describe the y-axis (Average)
- Displaying the plot on the screen


```python

df.plot(kind='bar',figsize=(22, 8))
plt.xticks(rotation='horizontal')
plt.title('Avg. HP,Defence,Attack,etc of each type of pokemon',size=20)
plt.xlabel('Type of pokemon')
plt.ylabel('Avg')
plt.show()
```

selecting all the weakness columns

```python
weakness_colms= [ 'Weakness-Normal','Weakness-Fire', 'Weakness-Water', 'Weakness-Electric',
       'Weakness-Grass', 'Weakness-Ice', 'Weakness-Fighting',
       'Weakness-Poison', 'Weakness-Ground', 'Weakness-Flying',
       'Weakness-Psychic', 'Weakness-Bug', 'Weakness-Rock', 'Weakness-Ghost',
       'Weakness-Dragon', 'Weakness-Dark', 'Weakness-Steel', 'Weakness-Fairy']
```

The code calculates the mean values of all weakness of against each Pokémon type (Type 1) by grouping the data using the 'type1' column

```python
type_comparison = pokemon_data.groupby(by='Type 1').mean()[weakness_colms]
```

```python
type_comparison
```

Here we are ploting the heatmap which shows us the average weaknesses of Pokémon types

first line setting the figure size and the 2nd line we are using the seaborn's heatmap method to plot the heatmap.

```python
plt.figure(figsize=(14, 9))
sns.heatmap(data=type_comparison,annot=True)
```
