Let's consider a real-life example of data visualization using the following scenario:

Scenario: A retail store wants to analyze its monthly sales data for the last year (12 months) to identify sales trends and make informed decisions.


```python
'''Raw Data:

January: $15,000
February: $18,000
March: $20,500
April: $22,000
May: $19,500
June: $25,000
July: $28,000
August: $24,500
September: $23,000
October: $26,000
November: $30,000
December: $32,500'''
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,2))
# Sample data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [15000, 18000, 20500, 22000, 19500, 25000, 28000, 24500, 23000, 26000, 30000, 32500]

# Create a line plot
plt.bar(months, sales, color='#ee82ee')

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Data')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,2))
# Sample data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [15000, 18000, 20500, 22000, 19500, 25000, 28000, 24500, 23000, 26000, 30000, 32500]

# Create a line plot
plt.bar(months, sales, color='#ee82ee', width = 0.5)

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Data')

# for setting x lables vertical
plt.xticks(range(len(months)), months, rotation='vertical')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(5,2))

# Sample data
months = ['January', 'February', 'March', 'April', 'May', 'June', 'July', 'August', 'September', 'October', 'November', 'December']
sales = [15000, 18000, 20500, 22000, 19500, 25000, 28000, 24500, 23000, 26000, 30000, 32500]

# Create a line plot
plt.bar(months, sales, color='#ee82ee', width = 0.75)

# Add labels and title
plt.xlabel('Months')
plt.ylabel('Sales Amount ($)')
plt.title('Monthly Sales Data')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.show()
```

**Studentâ€™s Activity**:

Consider the monthly expenses of a household for various categories:


categories = ['Housing', 'Utilities', 'Transportation', 'Groceries', 'Entertainment', 'Healthcare', 'Education', 'Miscellaneous']

expenses = [1200, 300, 400, 500, 200, 150, 250, 100]

You can use this data to create a bar graph to visualize the distribution of expenses across these categories.

```python
import matplotlib.pyplot as plt

categories = ['Housing', 'Utilities', 'Transportation', 'Groceries', 'Entertainment', 'Healthcare', 'Education', 'Miscellaneous']
expenses = [1200, 300, 400, 500, 200, 150, 250, 100]

# Create a bar graph
plt.bar(categories, expenses, color='blue')

# Add labels and title
plt.xlabel('Categories')
plt.ylabel('Amount (USD)')
plt.title('Monthly Expenses of a Household')
#plt.title('Monthly Expenses of a Household \n This bar graph will visually represent the monthly expenses for each category,\n making it easy to compare and understand the distribution of expenditures within the household.')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Display the plot
plt.tight_layout()  # To avoid overlapping labels
plt.show()
#This bar graph will visually represent the monthly expenses for each category,
#making it easy to compare and understand the distribution of expenditures within the household.

```

```python
plt.figure(figsize=(5,10))
plt.suptitle('2 subplots')

sub=['maths', 'science', 'lang1', 'lang2']
m1=[70,90,50,60]
m2=[60,70,80,90]

plt.subplot(2, 1, 1)
plt.bar(sub, m1, color ='c')
plt.title('subplot 1')
plt.ylabel('marks')

plt.subplot(2, 1, 2)
plt.bar(sub, m2, color ='m')
plt.title('subplot 2')
plt.xlabel('subjects')
plt.ylabel('marks')

plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.suptitle('2 subplots')

sub=['maths', 'science', 'lang1', 'lang2']
m1=[70,90,50,60]
m2=[60,70,80,90]

plt.subplot(1, 2, 1)
plt.bar(sub, m1, color ='c')
plt.title('subplot 1')
plt.ylabel('marks')

plt.subplot(1, 2, 2)
plt.bar(sub, m2, color ='m')
plt.title('subplot 2')
plt.xlabel('subjects')
plt.ylabel('marks')

plt.show()
```

```python
import matplotlib.pyplot as plt

Activities	=['School','Study','Play',	'Sleep','Others']
Time_spent =[6, 4, 1.5, 8, 4.5]
colors = ['c', 'g', 'y','m', 'hotpink', '#32a8a8']

# Create a figure with subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(10, 8))

# Plot on each subplot
axs[0, 0].plot(Activities, Time_spent)
axs[0, 0].set_title('Subplot 1 line graph')

axs[0, 1].scatter(Activities, Time_spent)
axs[0, 1].set_title('Subplot 2 scatter graph')

axs[1, 0].bar(Activities, Time_spent)
axs[1, 0].set_title('Subplot 3 bar graph')

axs[1, 1].pie(Time_spent, labels=Activities, colors=colors,
        startangle=90,
        explode = (0, 0.1, 0.5,0.25,0.5),
        autopct = '%1.1f%%' )
axs[1, 1].set_title('Subplot 4 pie chart')

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()

```

```python
import matplotlib.pyplot as plt

# Create a figure with subplots
fig, axs = plt.subplots(nrows=2, ncols=2, figsize=(5, 4))

# Plot on each subplot
axs[0, 0].plot([1, 2, 3], [4, 5, 6])
axs[0, 0].set_title('Subplot 1')

axs[0, 1].scatter([2, 4, 6], [7, 3, 9])
axs[0, 1].set_title('Subplot 2')

axs[1, 0].bar(['A', 'B', 'C'], [10, 8, 12])
axs[1, 0].set_title('Subplot 3')

axs[1, 1].hist([5, 6, 7, 5, 8, 9, 7, 6, 7])
axs[1, 1].set_title('Subplot 4')

# Adjust layout
plt.tight_layout()

# Show the subplots
plt.show()
```
